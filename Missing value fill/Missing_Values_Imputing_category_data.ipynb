```python
dataset=dataset.fillna(10)  # all Null value convert 10
dataset.isnull().sum()
```

```python
dataset.info() #chek dataset stucture              object =category data            int,flot=numarical data
```

# backWord filling Foreord filling

```python
dataset=dataset.fillna(method="bfill")  # all  null value backword filling but last row null value not fill
print("null value=",dataset.isnull().sum().sum())
```

```python
dataset=dataset.fillna(method="ffill")    # all null values  forword filiing  but first row null value  not data fill 
print("null value=",dataset.isnull().sum().sum())
```

```python
# columns wise null value fill 

dataset=dataset.fillna(method="bfill",axis=1)  #backword null value fill 
dataset=dataset.fillna(method="ffill",axis=1)  # forword null value fill
```

```python
print("null value=",dataset.isnull().sum().sum())
```

```python
dataset.info()
```

```python
gender=dataset["Gender"].mode()[0]  # more use value show 
married=dataset["Married"].mode()[0]
print(gender)
print(married)
```

```python
dataset["Gender"]=dataset["Gender"].fillna(dataset["Gender"].mode()[0])   #column null value fill
dataset.isnull().sum() 
```

```python

dataset.select_dtypes(include="object").isnull().sum()       #object datatype column select
```

```python
dataset.select_dtypes(include="object").columns   #object type columns show 
```

```python
for i in dataset.select_dtypes(include="object").columns:    # object datatypes all calumns null value fill mode function
    dataset[i]=dataset[i].fillna(dataset[i].mode()[0])
```
