```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv(r"E:\tops\Data science\machine Learning\15 naive\Multiple_input.csv")
dataset.head(3)
```

```python
dataset=dataset.drop("Unnamed: 0",axis=1)
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="cgpa",y="iq",data=dataset,hue="placement")
plt.show()
```

```python
sns.kdeplot(data=dataset["iq"])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placement"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=44)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB ,BernoulliNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100,gnb.score(x_train,y_train)*100
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,gnb.predict(x_test))
```

```python
mb=MultinomialNB()
```

```python
mb.fit(x_train,y_train)
```

```python
mb.score(x_test,y_test)*100,mb.score(x_train,y_train)*100
```

```python
confusion_matrix(y_test,mb.predict(x_test))
```

```python
bnb=BernoulliNB()
```

```python
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100,bnb.score(x_train,y_train)*100
```

```python
confusion_matrix(y_test,bnb.predict(x_test))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from sklearn.metrics import  confusion_matrix
```

```python
confusion_matrix(y_test,lr.predict(x_test))
```
