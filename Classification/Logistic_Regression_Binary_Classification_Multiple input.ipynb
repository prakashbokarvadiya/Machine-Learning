```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset =pd.read_csv(r"E:\tops\Data science\practical\Classification\Multiple_input.csv")
dataset.head()
```

```python
dataset=dataset.drop(columns="Unnamed: 0")
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="cgpa",y="iq",data=dataset,hue="placement")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```

```python
sns.scatterplot(x="cgpa",y=lr.predict(x),data=dataset,hue="placement",c="red")
plt.show()

```
