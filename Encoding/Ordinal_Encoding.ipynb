# scikit learn library through Encoding OrdinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord1=OrdinalEncoder(categories=ar)   #ordinal Encoder coll
```

```python
ord1.fit(dataset[["Property_Area"]])   #model fit
```

```python
lt=ord1.fit_transform(dataset[["Property_Area"]]) # model fit transfor both tugather
```

```python
lt
```

```python
type(lt)
```

```python
dataset["Property_Area"]=lt                      #origanl data set change encoding
```

# map function through encoding

```python
l1=list(dataset["Property_Area"].unique())                   # list 1 create 
l2=[]
l2=list((i for i in range(len(l1))))                         # using list comprasion list 2 create   
```

```python
data=dict(zip(l1,l2))                                      # list to dict convert 
```

```python
data
```

```python
dataset["Property_Area"]=dataset["Property_Area"].map(data)          #origanal data in change 
```
