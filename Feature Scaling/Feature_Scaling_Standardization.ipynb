```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv(r"E:\tops\Data science\practical\loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["LoanAmount"].fillna(dataset["LoanAmount"].mean(),inplace=True)
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
ss= StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

Save in dataset

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome"])
plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["ApplicantIncome_ss"])
plt.show()
```
