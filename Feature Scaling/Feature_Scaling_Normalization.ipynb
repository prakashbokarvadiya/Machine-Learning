```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset= pd.read_csv(r"E:\tops\Data science\practical\loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["LoanAmount"].fillna(dataset["LoanAmount"].mean(),inplace=True)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.title("ApplicantIncome")
sns.distplot(dataset["ApplicantIncome"])
plt.subplot(1,3,2)
plt.title("CoapplicantIncome")
sns.distplot(dataset["CoapplicantIncome"])
plt.subplot(1,3,3)
plt.title("LoanAmount")
sns.distplot(dataset["LoanAmount"])
plt.show()
```

```python
dataset.info()
```

```python
ms=cl=la = MinMaxScaler()
ms.fit(dataset[["ApplicantIncome"]])
cl.fit(dataset[["CoapplicantIncome"]])
la.fit(dataset[["LoanAmount"]])
```

```python
dataset[["ApplicantIncome_min"]] = ms.fit_transform(dataset[["ApplicantIncome"]])
dataset[["CoapplicantIncome_min"]] = cl.fit_transform(dataset[["CoapplicantIncome"]])
dataset[["LoanAmount_min"]]=la.fit_transform(dataset[["LoanAmount"]])
#ms.transform(dataset[["ApplicantIncome"]])
# ms.transform(dataset[["ApplicantIncome"]])
# cl.transform(dataset[["CoapplicantIncome"]])
#la.transform(dataset[["LoanAmount"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["ApplicantIncome_min"])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_min"])

plt.show()
```
