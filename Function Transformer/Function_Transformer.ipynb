```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from  sklearn.preprocessing import FunctionTransformer
```

```python
dataset =pd.read_csv(r"E:\tops\Data science\practical\loan.csv")
dataset.head(3)
```

```python
plt.subplot(1,1,1)
sns.distplot(x= dataset["CoapplicantIncome"])
plt.show()
```

```python
q1=dataset["CoapplicantIncome"].quantile(0.25)
q3=dataset["CoapplicantIncome"].quantile(0.75)
IQR =q3-q1
min_=q1-(1.5*IQR)
max_=q3+(1.5*IQR)
```

```python
min_,max_
```

```python
newdataset=dataset[dataset["CoapplicantIncome"]<=max_]
```

```python
newdataset.shape
```

```python
dataset=dataset[dataset["CoapplicantIncome"]<max_]
```

```python
dataset.shape
```

```python
ft =FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_ft"]=ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_ft"])
```
